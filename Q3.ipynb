```python
from selenium.webdriver.common.by import By
from selenium.webdriver.common.action_chains import ActionChains
import pandas as pd
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.common.by import By
from selenium.webdriver.support import expected_conditions as EC
from pyquery import PyQuery as pq
import time
import random
```

```python
def generate_news(html):
    doc = pq(html)#获取网页源代码
    res =[]#储存提取的信息
    items = doc('.news-item').items()#定位新闻框
    for item in items:
        news_title = item.find('h3 a').text()#获取新闻标题
        news_herf = item.find('h3 a').attr('href')#获取新闻链接
        news_content = item.find('.news-desc').text()[:-3]#获取新闻内容
        res.append([news_title,news_herf,news_content])
    return res
```

```python
def news_information(html):
    doc = pq(html)#获取网页源代码
    res =[]#储存提取的信息
    items = doc('.newsContent').items()#定位新闻框
    for item in items:
        news_time = item.find('.time').text()#获取新闻时间
        news_source = item.find('.source a').text()#获取新闻来源
        res.append([news_time,news_source])
    return res
```

```python
options = webdriver.ChromeOptions()#创建自动化测试选项
options.add_experimental_option('excludeSwitches', ['enable-automation'])#设置初始参数
driver = webdriver.Chrome(chrome_options=options)#传入自动化测试option
driver.maximize_window()#最大化窗口
all_herf_list = []#储存新闻标题和链接的列表
for i in range(10):
    driver.get('http://so.eastmoney.com/news/s?keyword=%E6%95%B0%E6%8D%AE%E6%B8%AF&pageindex='+str(i+1))#输入网址
    time.sleep(0.3)#休息0.3秒防止出问题
    html =driver.page_source#获得网页源代码
    all_herf_list += generate_news(html)#获得新闻标题和链接的列表
```

```python
#进入每个新闻链接并获取信息
for i in range(len(all_herf_list)):
    driver.get(all_herf_list[i][1])
    html =driver.page_source#获得网页源代码
    content_list = news_information(html)
    all_herf_list[i].append(content_list[0][0])
    all_herf_list[i].append(content_list[0][1])
```

```python
file=open('Q3_Answer.txt','w',encoding='UTF-8') 
for i in all_herf_list:
    file.write(str(i)+"\n")
file.close() 
```
